```
import numpy as np
import math
```

```
# Initialization
X = np.array([(0,0),(1,2),(1,3),(2,2),(2,3),(3,3),(6,8),(4,8)])
k = 3
X.shape, k
```

**Step 1 : Finding the distance between every pair or coordinates**

```
# This function finds out the Manhattan distance between the a and b coordinates
def dist(a, b):
  return np.sum(np.abs(a-b))
```

```
dis = []

for i in range(X.shape[0]):
  cur = []
  for j in range(X.shape[0]):
    cur.append(dist(X[i],X[j]))
  cur = np.asarray(cur)
  dis.append(cur)

dis = np.asarray(dis)
dis
```

**Step - 2 : Finding the k-th nearest neighbor distance from each coordinate**

```
knnd = []

for i in range (dis.shape[0]):
  cur = dis[i]
  cur = np.sort(cur)
  knnd.append(cur[3])

knnd = np.asarray(knnd)
knnd
```

**Step - 3 : Finding the k distance neighborhood of each coordinate**

```
knn = []

for i in range (dis.shape[0]):
  cn = []
  for j in range (dis.shape[1]):
    if knnd[i] >= dis[i][j] and i != j :
      cn.append(j)
  cn = np.asarray(cn)
  knn.append(cn)

knn = np.asarray(knn)
knn
```

**Step- 4 : Calculating local reachability distance for each coordinate**

```
# reachdist(a,b) : reachability distance from b to a
def reachdist(a,b):
  return max(knnd[b],dis[a][b])
```

```
lrd = []

for i in range (knn.shape[0]):
  sum = 0
  for j in range (knn[i].shape[0]):
    sum += reachdist(i,knn[i][j])
  cr = knn[i].shape[0]/sum
  lrd.append(cr)

lrd = np.asarray(lrd)
lrd
```

**Step- 5 : Calculating local outlier factor (LOF) for each coordinate**

```
LOF = []

for i in range(lrd.shape[0]):
  sum = 0
  for j in range(knn[i].shape[0]):
    sum += lrd[knn[i][j]]/lrd[i]
  cl = sum/knn[i].shape[0]
  LOF.append(cl)

LOF = np.asarray(LOF)
LOF
```

**Step - 6 : Identifying potential outliers**

```
outliers = []

for i in range(LOF.shape[0]):
  if LOF[i] >= 1.5 :
    outliers.append(X[i])

outliers = np.asarray(outliers)
outliers
```
